```python
scene_path = os.path.join('', 'scenes') 
results_spatio_path = os.path.join('','results_spatio') 
mi.set_variant('scalar_rgb')
texture_path = os.path.join(scene_path,'textures')

co_box = os.path.join(scene_path,'cornell-box/scene.xml')
che = os.path.join(scene_path,'110935.png')
```

```python
scene1 = mi.load_dict({
    'type':'scene',
    'integrator': {'type': 'direct'},
    'light': {
        'type': 'projector',
        'irradiance': {
            'type': 'rgb',
            #amount of light emitted
            'value': 0.5
        },
        #field of view
        'fov': 45,
        #mi.ScalarTransform4f.look_at
        'to_world': mi.ScalarTransform4f.look_at(
                origin=[-0.03, 0.1, -0.7],
                target=[-0.03, -0.1, -15],
                up=[0, 0, 1]
        )
    },
    # 'tok':{
    #     'type': 'ply',
    #     'filename': tokki,
    #     'to_world': mi.ScalarTransform4f.translate([0, 0, -1.0]),
    #     #Bidirectional Scattering Distribution Function
    #     'bsdf': {
    #         #texture
    #         'type': 'diffuse',
    #         'reflectance': {'type': 'rgb', 'value': [0.9, 0.2, 0.3]},
    #     }
    # }
    'rectangle' : {
        'type': 'rectangle',
        'to_world': mi.ScalarTransform4f.translate([0, 0.3, -4.0]),  # Modify rotation and translation here
        'bsdf': {
            'type': 'diffuse',
            'reflectance': {
                'type': 'bitmap',
                'filename': che
            }
        }
    }
    'environment': {
        'type': 'constant',
        'radiance': {
            'type': 'rgb',
            'value': [0.2, 0.4, 0.8]  # RGB values of the new background color
        }
    }
})


cam1 = mi.load_dict({
    'type': 'perspective',
    'fov': 45, #24.6 mm
    'to_world': mi.ScalarTransform4f.look_at(
        origin=[0.03, 0, -0.7],  # increase the z-coordinate to move the camera farther away
        target=[0.03, 1, -15],
        up=[0, 0, 1]
    ),
    'film': {
        'type': 'hdrfilm',
        'width': 360,
        'height': 360,
        
    }
# 'distortion': [0, 0, 0, 0, 0]
})


image = mi.render(scene=scene1, sensor=cam1, spp=64)
mi.Bitmap(image)
```
