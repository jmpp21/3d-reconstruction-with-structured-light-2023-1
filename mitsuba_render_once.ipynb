```python
# <bsdf texture type>

# diffuse: A matte surface that scatters incoming light uniformly in all directions.

# specular: A mirror-like surface that reflects incoming light in a single direction.

# dielectric: A surface that separates two different mediums (e.g. air and water), and can reflect and refract light based on the refractive index of the two mediums.

# conductor: A surface that conducts electricity, such as metals, which reflects and absorbs light in a way that depends on the frequency of the light.

# roughconductor: A rough version of the conductor BSDF, which accounts for the microstructure of real-world metals.

# roughdielectric: A rough version of the dielectric BSDF, which accounts for the surface roughness of materials like glass and plastic.

# difftrans: A combination of a diffuse and a dielectric BSDF, which allows light to both refract and scatter through a material, as in frosted glass.

# ward: A BSDF that models anisotropic surfaces, which have different reflective properties depending on the viewing angle.

```

```python
scene_path = os.path.join('', 'scenes') 
results_spatio_path = os.path.join('','results_spatio') 
mi.set_variant('scalar_rgb')
texture_path = os.path.join(scene_path,'textures')

co_box = os.path.join(scene_path,'cornell-box/scene.xml')
tokki = os.path.join(scene_path,'meshes/bunny.ply')
```

```python
scene1 = mi.load_dict({
    'type':'scene',
    'integrator': {'type': 'direct'},
    'light': {
        'type': 'projector',
        'irradiance': {
            'type': 'rgb',
            #amount of light emitted
            'value' : 0.5
        },
        #field of view
        'fov': 45,
        #mi.ScalarTransform4f.look_at
        'to_world': mi.ScalarTransform4f.look_at(
            #the position of light(x, y, z)
            #x: horizontal, y:vertical, z:depth
            origin=[-0.03, 0, -0.7],
            #light pointing at
            target=[-0.03, 1, -15],
            #
            up=[0, 0, 1]
        )
        },
    'tok':{
        'type': 'ply',
        'filename': tokki,
        'to_world': mi.ScalarTransform4f.translate([0, 0, -1.0]),
        #Bidirectional Scattering Distribution Function
        'bsdf': {
            #texture
            'type': 'diffuse',
            'reflectance': {'type': 'rgb', 'value': [0.9, 0.2, 0.3]},
        }
    }
})
cam1 = mi.load_dict({
    'type': 'perspective',
    'fov': 45, #field of view
    'to_world': mi.ScalarTransform4f.look_at(
        origin=[0.03, 0, -0.7],
        target=[0.03, 1, -15],
        up=[0, 0, 1]
    ),
    'film2': {
        'type': 'hdrfilm',
        'width': 360,
        'height': 360,
    }
})
image3 = mi.render(scene=scene1, sensor=cam1, spp=64)
mi.Bitmap(image3)
```
